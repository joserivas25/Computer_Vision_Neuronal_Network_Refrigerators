```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import torch
import os
import uuid
from pyspark.sql import Row
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, IntegerType
import shutil
from concurrent.futures import ThreadPoolExecutor, as_completed
import multiprocessing
from IPython.display import display as ipy_display, Image as IPImage
```

```python
dbutils.fs.ls("/mnt/canlusneblobstrg/")

# Define the path of the CSV file in the Blob Storage using Databricks file system utilities
images_path = "/mnt/canlusneblobstrg/cia_Postobon/vp_Ventas/Dir_PlaneacionVentas/jgrivas/02_OUTPUT/CNN Neveras/Fotos2/"

# Cargando el modelo desde archivo
modelo_path = "/Workspace/Users/jgrivas@postobon.com.co/yolo11s.pt"
model = YOLO(modelo_path)
```

```python
# -------------------------------------------
# 2. FunciГіn para procesar una imagen y devolver resultados
# -------------------------------------------


def obtener_color_promedio_y_nombre(imagen_path, box):
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    imagen = Image.open(imagen_path).convert("RGB")
    cropped = imagen.crop((x1, y1, x2, y2))
    pixels = np.array(cropped).reshape(-1, 3)
    rgb_prom = np.mean(pixels, axis=0).astype(int).tolist()

    colores_base = {
        # Tonos de rojo
        "rojo claro": (255, 102, 102),
        "rojo": (255, 0, 0),
        "rojo oscuro": (139, 0, 0),
        "vino tinto": (128, 0, 32),
        "granate": (153, 0, 0),

        # Tonos de azul
        "azul claro": (173, 216, 230),
        "azul": (0, 0, 255),
        "azul oscuro": (0, 0, 139),
        "azul navy": (0, 0, 128),

        # Tonos de rosado
        "rosado claro": (255, 192, 203),
        "rosado": (255, 105, 180),
        "rosado fuerte": (255, 20, 147),
        "fucsia": (255, 0, 255),

        # Tonos de morado
        "lavanda": (230, 230, 250),
        "morado claro": (186, 85, 211),
        "morado": (128, 0, 128),
        "pГєrpura": (160, 32, 240)
    }

    def distancia(c1, c2):
        return np.linalg.norm(np.array(c1) - np.array(c2))

    color_nombre = min(colores_base.items(), key=lambda kv: distancia(rgb_prom, kv[1]))[0]
    return rgb_prom, color_nombre



def detectar_neveras(imagen_path):
    results = model(imagen_path)
    detecciones = []

    id_imagen = str(uuid.uuid4())  # Гєnico por imagen

    for r in results:
        for box in r.boxes:
            clase = int(box.cls[0])
            nombre_clase = model.names[clase]
            conf = float(box.conf[0])
            x1, y1, x2, y2 = map(float, box.xyxy[0])

            rgb_promedio, color_nombre = obtener_color_promedio_y_nombre(imagen_path, box)

            detecciones.append(Row(
                id_imagen=id_imagen,
                nombre_archivo=os.path.basename(imagen_path),
                clase=nombre_clase,
                confianza=conf,
                x1=x1,
                y1=y1,
                x2=x2,
                y2=y2,
                rgb_promedio=str(rgb_promedio),
                color=color_nombre
            ))
    return detecciones    
```

```python
# -------------------------------------------
# 3. Obtener lista de imГЎgenes
# -------------------------------------------

image_files = [
    f"/dbfs{images_path}{f}" 
    for f in os.listdir(f"/dbfs{images_path}") 
    if f.lower().endswith((".jpg", ".png", ".jpeg"))
]

# -------------------------------------------
# 4. Procesamiento paralelo con CPU
# -------------------------------------------

num_hilos = multiprocessing.cpu_count()
print(f"Procesando con {num_hilos} hilos...")

todos_los_resultados = []

with ThreadPoolExecutor(max_workers=num_hilos) as executor:
    futuras = {executor.submit(detectar_neveras, path): path for path in image_files}
    for future in as_completed(futuras):
        path = futuras[future]
        try:
            detecciones = future.result()
            todos_los_resultados.extend(detecciones)
        except Exception as e:
            print(f"Error procesando {path}: {e}")

# -------------------------------------------
# 5. Crear carpeta para resultados visuales
# -------------------------------------------

visual_output_path = "/dbfs/tmp/neveras_viz/"
local_output_path = "/tmp/neveras_viz/"
shutil.rmtree(local_output_path, ignore_errors=True)
os.makedirs(local_output_path, exist_ok=True)

# -------------------------------------------
# 6. Guardar imГЎgenes con detecciones (SECUENCIAL)
# -------------------------------------------

for img_path in image_files[:500]:
    results = model(img_path, save=True, save_dir=local_output_path)

    archivo = os.path.basename(img_path)
    resultado_path = os.path.join(local_output_path, "predict", archivo)

    if os.path.exists(resultado_path):
        ipy_display(IPImage(filename=resultado_path))
```

```python
from pyspark.sql.types import *
from pyspark.sql.functions import col, regexp_replace

# -------------------------------------------
# 7. Guardar CSV con resultados en Blob Storage
# -------------------------------------------

csv_output_path = (
    "/mnt/canlusneblobstrg/cia_Postobon/vp_Ventas/Dir_PlaneacionVentas/jgrivas/02_OUTPUT/CNN Neveras/neveras_detecciones2.csv"
)

spark.conf.set("spark.databricks.delta.formatCheck.enabled", "false")

schema = StructType([
    StructField("id_imagen", StringType(), True),
    StructField("nombre_archivo", StringType(), True),
    StructField("clase", StringType(), True),
    StructField("confianza", DoubleType(), True),
    StructField("x1", DoubleType(), True),
    StructField("y1", DoubleType(), True),
    StructField("x2", DoubleType(), True),
    StructField("y2", DoubleType(), True),
    StructField("rgb_promedio", StringType(), True),
    StructField("color", StringType(), True)
])

df_resultados = spark.createDataFrame(todos_los_resultados, schema)

# Reemplazar punto por coma en columnas numГ©ricas
df_final = df_resultados.withColumn("confianza", regexp_replace(col("confianza").cast("string"), r"\.", ",")) \
                        .withColumn("x1", regexp_replace(col("x1").cast("string"), r"\.", ",")) \
                        .withColumn("y1", regexp_replace(col("y1").cast("string"), r"\.", ",")) \
                        .withColumn("x2", regexp_replace(col("x2").cast("string"), r"\.", ",")) \
                        .withColumn("y2", regexp_replace(col("y2").cast("string"), r"\.", ","))

df_final.createOrReplaceTempView("neveras_detecciones")

df_final.write.mode("overwrite").option("header", True).csv(csv_output_path)
```
